```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Group4_0001_6D.tsv',sep='\t',skiprows=11)
df
```

```python
df.columns
```

```python
gimbal_pos = df[['Gimbal1 X','Y','Z']].to_numpy()
gimbal_rpy = df[['Roll','Pitch','Yaw']].to_numpy()
drone_pos = df[['M4 X','Y.2','Z.2']].to_numpy()
drone_rpy = df[['Roll.2','Pitch.2','Yaw.2']].to_numpy()
```

```python
gimbal_pos[0]
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection = '3d')
# ax.plot(gimbal_pos[:,0], gimbal_pos[:,1], gimbal_pos[:,2], marker = '.')
ax.plot(gimbal_pos[:,0], gimbal_pos[:,1], gimbal_pos[:,2])
ax.plot(drone_pos[:,0], drone_pos[:,1], drone_pos[:,2])
plt.show()
```

```python
fig, axs = plt.subplots(3,2, figsize=(12,8))

labels = ['x','y','z']
for i in range(3):
    axs[i,0].plot(drone_pos[:,i], label=f'drone-{labels[i]}')
    axs[i,0].plot(gimbal_pos[:,i], label=f'gimbal-{labels[i]}')
    axs[i,0].legend()

labels = ['roll', 'pitch', 'yaw']
for i in range(3):
    axs[i,1].plot(drone_rpy[:,i], label=f'drone-{labels[i]}')
    axs[i,1].plot(gimbal_rpy[:,i], label=f'gimbal-{labels[i]}')
    axs[i,1].legend()

plt.show()
```
